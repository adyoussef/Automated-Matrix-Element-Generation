## Computing the Scattering Cross-Section

```python
me = 0.000511
mp = 0.938
ee = (me**2 + 100**2)**0.5
ep = (mp**2 + 100**2)**0.5
pp1 = ps2.FourVector(ee,0,0,100)
pp2 = ps2.FourVector(ep,0,0,-100)

process = Process("e+ e- -> mu+ mu-")

h = [None,None,None,None]

start_time = time.time()
s = Scattering(process,pp1,pp2,h, frame ="jo")
i = ps2.Integrator(s.dxs, 0.0, 2*pi, 0, pi)

print(i.mc(1000)/1e-31)
```
